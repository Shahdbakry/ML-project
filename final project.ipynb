```python
import pandas as pd
data=pd.read_csv('https://storage.googleapis.com/kagglesdsdata/datasets/74977/169835/StudentsPerformance.csv?X-Goog-Algorithm=GOOG4-RSA-SHA256&X-Goog-Credential=gcp-kaggle-com%40kaggle-161607.iam.gserviceaccount.com%2F20230526%2Fauto%2Fstorage%2Fgoog4_request&X-Goog-Date=20230526T223537Z&X-Goog-Expires=259200&X-Goog-SignedHeaders=host&X-Goog-Signature=3721faa6ab90f461e2ead2e4c4c05211b156adc4616c6f9495043255d7e50b6cc1a9e41b140d2331f3dbfd4e012b816d2376532b335f48329dfc65565c6379eec27193a7a8c14dde346f32e5fdbfe13c3eafb27bc4513670e69fb44ac35288e626fa42eb10d2fbbe10f9588657fe434efd68de523ad8b6e3a13ef064d80d2c0831dbcb436704642fe371b1914adb65d5a468d2b9050f1fd59c14bb76ee0620ad28daf1fd88ce27c703e0823d5a561840865e2d910099943c922fe6af89a038e2e156976395d218a359e53c8768b2901cc8ae5c9bc7a8fdd33377cdbdab08e2d95e75221ff065433486bb13b5e47845f011be4ea109e1036bae8112b083c1f097')
# print(data)

data_cat=data.select_dtypes(["object"])

from sklearn import preprocessing
dtypes=data_cat.dtypes
for i in range (data_cat.shape[1]):
    if dtypes[i]=='object':
        Pr_data = preprocessing.LabelEncoder()
        data_cat[data_cat.columns[i]]=Pr_data.fit_transform(data_cat[data_cat.columns[i]])
        # print("Column index = ", i)
        # print(Pr_data.classes_)

X=data_cat.iloc[:,:-1].values
Y=data_cat.iloc[:,-1].values
# print(X)
# print(Y)

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size= 0.3)
# print("x_Train = ",X_train)
# print("x_test=",X_test) 
# print("y_train=",Y_train)
# print(" y_test=",Y_test)

# hen we should train our model by using the dataset. We will select Support
# Vector Machine as a suitable AI technique for our dataset as follows:

from sklearn.svm import SVC
clf = SVC(kernel='linear')
clf.fit(X_train, Y_train)
Accuracy=clf.score(X_test,Y_test)
print("Accuracy = ",Accuracy)

# After split the data into train & test we shouhd know the value of error that we have in this data
# So we use mean absolute error to know it and get the y predict from x_test

Y_pred=clf.predict(X_test)
from sklearn import metrics
MAE=metrics.mean_absolute_error(Y_test,Y_pred)
print('Mean Absolute Error = ', MAE)

# To display ROC curve, we will import “RocCurveDisplay” from
# “sklearn.metrics” library. Then we will use “.from_estimator”
# And the AUC is calculated on the curve which is equal  0.51

from sklearn.metrics import RocCurveDisplay
RocCurveDisplay.from_estimator(clf, X_test, Y_test)
import matplotlib.pyplot as plt
plt.show()

#  to determine the performance of the classification models for a given set of test data.
# For binary classification, confusion matrix consists from two rows and two columns. The
# two columns for true values while the rows for predicted values.

from sklearn.metrics import confusion_matrix
A = confusion_matrix(Y_test, Y_pred)
print(A)

# The cell values are: 
# • True Positive (TP) 
# • False Positive (FP) (Type-I error) 
# • False Negative (FN) (Type-II error) 
# • True Negative (TN)
# FP and FN occur when mismatching between the real value and predicted value.
# So, they are error. FP called is Type-1 error while FN is called Type-2 error.

TN, FP, FN, TP = confusion_matrix(Y_test, Y_pred).ravel()
print('TN = ', TN)
print('FP = ', FP)
print('FN = ', FN)
print('TP = ', TP)

# It can be defined as the number of correct outputs provided by the model
# or out of all positive classes that have predicted correctly by the model,
# how many of them were actually true.

P = metrics.precision_score(Y_test, Y_pred)
print("Precision score = ",P)

# Recall:
# It is defined as the out of total positive classes, how our model predicted
# correctly. The recall must be as high as possible.

R = metrics.recall_score(Y_test, Y_pred)
print("Recall score = ",R)

# It defines how often the model predicts the correct output.
# It can be calculated as the ratio of the number of correct predictions
# made by the classifier (TP + TN) to all number of predictions made by the classifiers.

Acc=metrics.accuracy_score(Y_test, Y_pred)
print("Recall score = ",Acc)

# f1-measure (F-Score or F1-Score):
# If two models have low precision and high recall or vice versa, it is difficult to
# compare these models. So, for this purpose, we can use F-score. This score
# helps us to evaluate the recall and precision at the same time. The F-score is
# maximum if the recall is equal to precision ('the output shown this')

F1 = metrics.f1_score(Y_test, Y_pred)
print("F1 score = ",F1)
```
